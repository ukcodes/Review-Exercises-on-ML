# Exercise 2 - Visualize data

```python
# visualize the sepal length and sepal width in a scatter plot
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
# the mspatches is required to display the plot
x = iris.data[:, 0] # x-axis - sepal length
y = iris.data[:, 1] # y-axis - sepal width
species = iris.target # species
x_min, x_max = x.min()-0.5, x.max()+0.5
y_min, y_max = y.min()-0.5, y.max()+0.5
plt.figure()
plt.title('Iris Dataset - Classification by Sepal Sizes')
plt.scatter(x, y, c = species)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
# the black ones are Iris setosa, green ones are Iris versicolor, and the yellow ones are Iris virginica
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
# the mspatches is required to display the plot
x = iris.data[:, 2] # x-axis - petal length
y = iris.data[:, 3] # y-axis - pepal width
species = iris.target # species
x_min, x_max = x.min()-0.5, x.max()+0.5
y_min, y_max = y.min()-0.5, y.max()+0.5
plt.figure()
plt.title('Iris Dataset - Classification by Petal Sizes', size = 14)
plt.scatter(x, y, c = species)
plt.xlabel('Petal length')
plt.ylabel('Petal width')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
plt.show()
```

# Exercise 3 - Principal Component Analysis (PCA)

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
x_reduced = PCA(n_components = 3).fit_transform(iris.data)
fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('Iris Dataset by PCA', size = 14)
ax.scatter(x_reduced[:,0], x_reduced[:,1], x_reduced[:, 2], c=species)
ax.set_xlabel('First eigenvector')
ax.set_ylabel('Second eigenvector')
ax.set_zlabel('Third eigenvector')
ax.w_xaxis.set_ticklabels(())
ax.w_yaxis.set_ticklabels(())
ax.w_zaxis.set_ticklabels(())
```

# Exercise 4- Implementation of KNN classification

```python
# KNN
# randomnly shuffle the data and divide it into train and test data
import numpy as np
np.random.seed(0)
x = iris.data
y = iris.target
i = np.random.permutation(len(x))
x_train = x[i[:-10]]
y_train = y[i[:-10]]
x_test = x[i[-10:]]
y_test = y[i[-10:]]
```

```python
# KNN
# train the KNN classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
#note - PCA performed earlier has not been used in the KNN classification
```

```python
knn.predict(x_test)
```

# Exercise 5a- Visualization of decision boundaries

```python
# KNN
# classify based on sepal length and width; visualize the decision boundaries
# This exercise is not a continuation of the previous exercise in which the data was shuffled and the KNN was trained using 
# train data. In this exercise, all the sepal length and width data are used to make the decision boundaries and visualize
# them in a scatter plot of sepal length (x-axis) vs sepal width (y-axis)
```

```python
from matplotlib.colors import ListedColormap
x = iris.data[:,:2] # x-axis - sepal length and width
y = iris.target # y-axis - species
x_min, x_max = x[:,0].min() - 0.5, x[:,0].max()+0.5
y_min, y_max = x[:, 1].min() - 0.5, x[:,1].max()+0.5

# MESH
cmap_light = ListedColormap(['#AAAAFF', '#AAFFAA', '#FFAAAA'])
h = 0.02
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
knn.fit(x, y)
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap = cmap_light)

# plot the training points
plt.scatter(x[:,0], x[:,1], c=y)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
```
